```python
import torch
from torchvision import datasets
from torchvision import transforms

import matplotlib.pyplot as plt
import numpy as np

import model
```

```python
device = torch.device("mps")

tensor_transform = transforms.ToTensor()
dataset = datasets.MNIST(root="~/pytorch/data", train=True, download=True, transform=tensor_transform)
```

```python

i = 0
_, axs = plt.subplots(nrows=4, ncols=4, figsize=(4, 4))
for i, ax in enumerate(axs.reshape(-1)):
    (data, target) = dataset[i]
    data = data.reshape((28, 28))

    ax.set_title(target)
    ax.imshow(data, cmap=plt.cm.gray, origin="upper", interpolation="nearest")
    ax.axis("off")
    i += 1

plt.tight_layout()
plt.show()
```
